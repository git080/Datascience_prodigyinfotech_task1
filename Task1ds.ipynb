```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"C:\\Users\ACER\Downloads\\task1\world_population.csv") 
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='1960')
plt.title('Histogram of 1960')
plt.xlabel('1960')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country Name', y='1960')
plt.title('Barplot for the Year 1960')
plt.xlabel('Country Name')
plt.ylabel('Value')
plt.show() 
```
